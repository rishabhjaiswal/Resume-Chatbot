**clean_text** function that does these things in order:
1.   Lowercase everythng
2.   Remove punctuation and numbers
Split text into individual words (tokenize)
Remove common filler words like "the", "and", "is" (stopwords)
Reduce words to their root form — "running" becomes "run"
(lemmatization)



```python
from os.path import join
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')


```

```python
def clean_text(text):
  lower_text=text.lower()
  remove_punctuation=''.join(letter for letter in lower_text if letter.isalnum() or letter.isspace()) # Allow spaces
  remove_numbers=''.join(letter for letter in remove_punctuation if not letter.isdigit())
  tokens = word_tokenize(remove_numbers)
  filtered_words = [word for word in tokens if word not in stopwords.words('english')]
  lemmatizer = WordNetLemmatizer()
  lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]
  return lemmatized_words

print(clean_text("The cats were running faster than the dogs."))
```

```python
skills_list = ["Python", "ML", "LLMs", "RAG", "PyTorch", "TensorFlow", "Hugging Face", "LangChain", "Vector Databases", "MLOps", "Docker", "Kubernetes"]
my_skills = ["Machine Learning", "Deep Learning", "Python", "Large Language Models (LLMs)", "Generative AI", "Retrieval-Augmented Generation (RAG)", "Prompt Engineering", "LangChain", "LlamaIndex", "Vector Databases", "Fine-Tuning", "Hugging Face", "PyTorch", "TensorFlow", "MLOps", "Model Deployment", "Model Evaluation", "NLP (Natural Language Processing)", "SQL", "Docker / Kubernetes"]

def extract_keywords(skills):
  return {skill for skill in skills if skill in skills_list}

result = extract_keywords(my_skills)


print(result)
```
